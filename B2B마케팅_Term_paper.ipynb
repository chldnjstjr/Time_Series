```r
naver_ts <- ts(naver)
naver_ts
```

```r
str(naver_ts)
```

```r
t <- naver$DATE
s <- naver$SVI
```

```r
plot(naver_ts)
```

```r
# 차분을 통해 데이터 정상화
naver_diff1 <- diff(s, differences = 1)
naver_diff2 <- diff(s, differences = 2)
naver_diff3 <- diff(s, differences = 3)

plot.ts(naver_diff1)    # 1차 차분만 해도 어느정도 정상화 패턴을 보임
plot.ts(naver_diff2)
plot.ts(naver_diff3)
```

```r
mean(naver_diff3); sd(naver_diff3)
```

```r
acf(naver_diff3, lag.max = 20, main="Autocorrelation Function")      # lag 2부터 점선 안에 존재. lag 절단값 = 2. --> MA(1)
```

```r
pacf(naver_diff3, lag.max = 20, main="Partial Autocorrelation Function")     # lag 5에서 절단값 --> AR(4)
```

```r

library(TTR)
library(forecast)
```

```r
auto.arima(s)
```

```r
arima(s, order = c(4,0,1))
model <- arima(s, order = c(4,0,1))
```

```r
install.packages("portes")
```

```r
library("portes")
```

```r
Box.test(model$residuals, lag=20, type="Ljung-Box") 
#시계얼 모형이 적합한지 알아보기 위해 Ljung-Box 검정을 실시한 결과, 유의확률은 0.000으로 유의수준 0.05하에서 통계적으로 유의한 것으로 나왔다. 
```

```r
cor(fitted(model), s)^2 #R제곱 계산
```
